```python
# # Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Renaming columns according to snake casing for better readability and documentation
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# create a new column age_group

labels = ['Young Adult','Adult','Middle_aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels)
```

```python
# --Q1. What is the total revenue generated by male vs. female customers?

query = """SELECT gender,SUM(purchase_amount) AS revenue
FROM name
GROUP BY gender
"""

df_result = pd.read_sql(query, conn)
print(df_result)
```

```python
# --Q2. Which customers used a discount but still spent more than the average purchase amount?
query2 = """select customer_id, purchase_amount
from name
where discount_applied = 'Yes' and purchase_amount >= (select AVG(purchase_amount) from name)"""

avg_purchese_amount = pd.read_sql(query2,conn)
avg_purchese_amount
```

```python
# -- Q3. Which are the top 5 products with the highest average review rating?
query3 = """ select item_purchased, round(avg(review_rating),2) as "Average Product Rating"
from name
group by item_purchased
order by "Average Product Rating" desc
limit 5 """

Avg_product_rating = pd.read_sql(query3,conn)
Avg_product_rating
```

```python
# --Q4. Compare the average Purchase Amounts between Standard and Express Shipping.
query4 = """select shipping_type, round(avg(purchase_amount),2)
from name
where shipping_type in ('Standard','Express')
group by shipping_type"""

Avg_purchase_amount = pd.read_sql(query4,conn)
Avg_purchase_amount
```

```python
# --Q5. Do subscribed customers spend more? Compare average spend and total revenue
# --between subscribers and non-subscribers.

query5 = """ select subscription_status,
  count(customer_id) as total_customers,
  round (avg(purchase_amount),2) as avg_spend,
  round(sum(purchase_amount),2) as total_revenue
from name
group by subscription_status
order by total_revenue desc , avg_spend desc"""

avg_spend = pd.read_sql(query5,conn)
avg_spend
# total_revenue = pd.read_sql()
```

```python
# --Q6. Which 5 products have the highest percentage of purchases with discounts applied?

query6 = """ SELECT item_purchased,
       ROUND(100.0 * SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END)/COUNT(*),2) AS discount_rate
FROM name
GROUP BY item_purchased
ORDER BY discount_rate DESC
LIMIT 5;"""

discount_rate = pd.read_sql(query6 , conn)
discount_rate
```

```python
# Q7. Segment customers into New, Returning, and Loyal based on their total number of previous purchases, and show the count of each segment.

query7 = """with customer_type as (
SELECT customer_id, previous_purchases,
CASE
    WHEN previous_purchases = 1 THEN 'New'
    WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
    ELSE 'Loyal'
    END AS customer_segment
FROM name)

select customer_segment,count(*) AS "Number of Customers"
from customer_type
group by customer_segment"""

cust_segment = pd.read_sql(query7 , conn)
cust_segment
```

```python
# --Q8. What are the top 3 most purchased products within each category?
query8 = """WITH item_counts AS (
    SELECT category,
           item_purchased,
           COUNT(customer_id) AS total_orders,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(customer_id) DESC) AS item_rank
    FROM name
    GROUP BY category, item_purchased
)
SELECT item_rank,category, item_purchased, total_orders
FROM item_counts
WHERE item_rank <=3"""

top_product = pd.read_sql(query8, conn)
top_product
```

```python
# Q9. Are customers who are repeat buyers (more than 5 previous purchases) also likely to subscribe?

query9 = """SELECT subscription_status,
       COUNT(customer_id) AS repeat_buyers
FROM name
WHERE previous_purchases > 5
GROUP BY subscription_status"""

repeat_buyers = pd.read_sql(query9, conn)
repeat_buyers
```

```python
# --Q10. What is the revenue contribution of each age group?
query10 = """SELECT age_group,
    SUM(purchase_amount) AS total_revenue
FROM name
GROUP BY age_group
ORDER BY total_revenue desc"""

total_revenue = pd.read_sql(query10, conn)
total_revenue
```
